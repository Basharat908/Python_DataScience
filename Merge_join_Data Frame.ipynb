How to perform a Merge Operation in DataFrame?

```python
import pandas as pd
player = ["Player1","Player2","Player3"]
points=[ 8,8,6]
title =["Game1","Game2","Game3"]
df = pd.DataFrame({"Player":player,"Points":points,"Title":title})
df
```

```python
player = [ "Player1","Basharat","javid"]
power = [ "Punch","kick", "Elbow"]
title = ["Game1","Game5","Game6"]
df2 = pd.DataFrame({"Player":player,"Power":power,"Title":title})
df2
```

```python
#INNER MERGE
df.merge(df2,on="Title", how = "inner")
```

```python
df.merge(df2,on = "Player",how =  "inner")
```

```python
#LEFT MERGE
df.merge(df2, on =  "Player", how = "left")
```

```python
#Rigt merge
df.merge(df2,on = "Player",how = "right")
```

```python
#Outer Merge
df.merge(df2,on = "Player",how = "outer")
```

How to do Join Operation in Pandas
Merge is based on attributes i.e it should have alteast one common column name but
in join operation is based on the index values when two tables have completely different values.

```python
player = [ "Basharat1","Javid5","Aabid6"]
power  = [ "Punch","Kick","Elbow"]
title  = [ "Game1","Game5","Game6"]
df3    = pd.DataFrame({"Player":player,"Power":power,"Title":title},index = ["L1","L2","L3"])
df3
```

```python
player = [ "Basharat1","Javid5","Aabid6"]
power  = [ "Punch","Kick","Elbow"]
title  = [ "Game1","Game5","Game6"]
df4    = pd.DataFrame({"Players":player,"Power1":power,"Titles":title},index = ["L2","L3","L4"])
df4
```

```python
#inner Join, Join is based on the smae number of columns 
df3.join(df4,how="inner")
```

```python
#Outer join! prints the all the rows and columns
df3.join(df4,how = "outer")
```

```python
#Left join!   all dataframes form left and common dataframes from right
df3.join(df4,how = "left")
```

```python
#Riht Join!  All dataframes from right but common dataframes from left
df3.join(df4, how = "right")
```

How to do concatinate two dfs in Pandas?

```python
pd.concat([df3,df4])
```
